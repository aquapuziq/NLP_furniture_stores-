```python
import pandas as pd 
import numpy as np
import sys
import requests
```

```python
file_path = r'C:\dev\projects\NLP_test\data\train_dataset_200.csv'
data = pd.read_csv(file_path)
```

```python
def url_is_available(url: str, timeout: int = 10) -> bool:
    try:
        r = requests.head(url, allow_redirects=True, timeout=timeout)
        if r.status_code < 400:
            return True
            
        r = requests.get(url, allow_redirects=True, timeout=timeout)
        return r.status_code < 400

    except requests.RequestException:
        return False
```

```python
urls = data['max(page)'].to_numpy()
```

```python
available = []

for url in urls:
    check = url_is_available(url)
    available.append(check)
```

```python
print(available)
```

```python
data['available'] = available
data.head()
```

```python
data.to_csv(r'C:\dev\projects\NLP_test\data\train_dataset_200_.csv', index=False, encoding='utf-8')
```
